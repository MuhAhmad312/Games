```python
import pygame
import random

# Initialize pygame
pygame.init()

# Constants
ROWS = 25
COLS = 50
TILE_SIZE = 25

WINDOW_WIDTH = TILE_SIZE * COLS
WINDOW_HEIGHT = TILE_SIZE * ROWS

# Colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 0, 0)
GREEN = (0, 255, 0)

# Tile class for representing snake segments and food
class Tile:
    def __init__(self, x, y):
        self.x = x
        self.y = y

# Snake class
class Snake:
    def __init__(self):
        self.x = TILE_SIZE * 5
        self.y = TILE_SIZE * 5
        self.velocityX = 1  # Initial velocity (moving right)
        self.velocityY = 0
        self.body = []  # List to hold segments of the snake's body
        self.score = 0
        self.game_over = False

    def change_direction(self, event):
        if self.game_over:
            return
        
        if event.key == pygame.K_UP and self.velocityY == 0:
            self.velocityX, self.velocityY = 0, -1
        elif event.key == pygame.K_DOWN and self.velocityY == 0:
            self.velocityX, self.velocityY = 0, 1
        elif event.key == pygame.K_LEFT and self.velocityX == 0:
            self.velocityX, self.velocityY = -1, 0
        elif event.key == pygame.K_RIGHT and self.velocityX == 0:
            self.velocityX, self.velocityY = 1, 0

    def move(self):
        if self.game_over:
            return
        
        # Update snake's body segments
        if len(self.body) > 0:
            for i in range(len(self.body) - 1, 0, -1):
                self.body[i].x = self.body[i - 1].x
                self.body[i].y = self.body[i - 1].y
            self.body[0].x = self.x
            self.body[0].y = self.y

        self.x += self.velocityX * TILE_SIZE
        self.y += self.velocityY * TILE_SIZE

        # Check if snake hits walls
        if self.x < 0 or self.x >= WINDOW_WIDTH or self.y < 0 or self.y >= WINDOW_HEIGHT:
            self.game_over = True
            return

        # Check if snake hits itself
        for segment in self.body:
            if (self.x, self.y) == (segment.x, segment.y):
                self.game_over = True
                return

        # Check if snake eats food
        if (self.x, self.y) == (food.x, food.y):
            self.score += 1
            self.body.append(Tile(food.x, food.y))
            place_food()

    def automatic_move(self):
        if self.game_over:
            return

        # Current and target positions
        head_x, head_y = self.x // TILE_SIZE, self.y // TILE_SIZE
        food_x, food_y = food.x // TILE_SIZE, food.y // TILE_SIZE

        # Calculate potential moves and prioritize safe ones
        potential_moves = []
        if self.velocityY != 1 and head_y > 0 and (head_x, head_y - 1) not in [(segment.x // TILE_SIZE, segment.y // TILE_SIZE) for segment in self.body]:
            potential_moves.append((0, -1))  # move up
        if self.velocityY != -1 and head_y < ROWS - 1 and (head_x, head_y + 1) not in [(segment.x // TILE_SIZE, segment.y // TILE_SIZE) for segment in self.body]:
            potential_moves.append((0, 1))  # move down
        if self.velocityX != 1 and head_x > 0 and (head_x - 1, head_y) not in [(segment.x // TILE_SIZE, segment.y // TILE_SIZE) for segment in self.body]:
            potential_moves.append((-1, 0))  # move left
        if self.velocityX != -1 and head_x < COLS - 1 and (head_x + 1, head_y) not in [(segment.x // TILE_SIZE, segment.y // TILE_SIZE) for segment in self.body]:
            potential_moves.append((1, 0))  # move right

        # Choose the move that gets closest to the food
        best_move = None
        min_distance = float('inf')
        for move in potential_moves:
            new_x, new_y = head_x + move[0], head_y + move[1]
            distance = abs(new_x - food_x) + abs(new_y - food_y)
            if distance < min_distance:
                min_distance = distance
                best_move = move

        if best_move:
            self.velocityX, self.velocityY = best_move

    def draw(self, screen):
        pygame.draw.rect(screen, GREEN, (self.x, self.y, TILE_SIZE, TILE_SIZE))
        for segment in self.body:
            pygame.draw.rect(screen, GREEN, (segment.x, segment.y, TILE_SIZE, TILE_SIZE))

# Food class
class Food:
    def __init__(self):
        self.x = TILE_SIZE * 10
        self.y = TILE_SIZE * 10

    def draw(self, screen):
        pygame.draw.rect(screen, RED, (self.x, self.y, TILE_SIZE, TILE_SIZE))

def place_food():
    food.x = random.randint(0, COLS - 1) * TILE_SIZE
    food.y = random.randint(0, ROWS - 1) * TILE_SIZE
    while (food.x, food.y) in [(snake.x, snake.y)] + [(segment.x, segment.y) for segment in snake.body]:
        food.x = random.randint(0, COLS - 1) * TILE_SIZE
        food.y = random.randint(0, ROWS - 1) * TILE_SIZE

# Initialize game variables
snake = Snake()
food = Food()

# Set up the display
screen = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption('Snake Game')

clock = pygame.time.Clock()

# Main game loop
while snake.score < 150:
    # Handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            quit()

    # Automatic movement logic for snake
    snake.automatic_move()
    snake.move()

    # Update the screen
    screen.fill(BLACK)
    snake.draw(screen)
    food.draw(screen)

    # Display score    
    font = pygame.font.SysFont('Arial', 20)
    score_text = font.render(f'Score: {snake.score}', True, WHITE)
    screen.blit(score_text, (10, 10))

    # Display game over message
    if snake.game_over:
        game_over_text = font.render(f'Game Over: {snake.score}', True, WHITE)
        screen.blit(game_over_text, ((WINDOW_WIDTH - game_over_text.get_width()) // 2, (WINDOW_HEIGHT - game_over_text.get_height()) // 2))

    pygame.display.flip()
    clock.tick(50)  # Adjust the speed of the game (10 FPS)

# Quit pygame properly
pygame.quit()
```

```python
import pygame
import random
import heapq

# Initialize pygame
pygame.init()

# Constants
ROWS = 25
COLS = 50
TILE_SIZE = 25

WINDOW_WIDTH = TILE_SIZE * COLS
WINDOW_HEIGHT = TILE_SIZE * ROWS

# Colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 0, 0)
GREEN = (255, 255, 255)

# Tile class for representing snake segments and food
class Tile:
    def __init__(self, x, y):
        self.x = x
        self.y = y

# Snake class
class Snake:
    def __init__(self):
        self.x = TILE_SIZE * 5
        self.y = TILE_SIZE * 5
        self.velocityX = 0
        self.velocityY = 0
        self.body = []  # List to hold segments of the snake's body
        self.score = 0
        self.game_over = False
        self.path = []
        self.path_index = 0

    def move(self):
        if self.game_over:
            return
        
        if self.path_index < len(self.path):
            next_step = self.path[self.path_index]
            self.x, self.y = next_step
            self.path_index += 1

        # Check if snake hits walls
        if self.x < 0 or self.x >= WINDOW_WIDTH or self.y < 0 or self.y >= WINDOW_HEIGHT:
            self.game_over = True
            return

        # Check if snake hits itself
        for segment in self.body:
            if (self.x, self.y) == (segment.x, segment.y):
                self.game_over = True
                return

        # Check if snake eats food
        if (self.x, self.y) == (food.x, food.y):
            self.score += 1
            self.body.append(Tile(food.x, food.y))
            place_food()
            self.path = find_safe_path((self.x, self.y), (food.x, food.y), self.body)
            self.path_index = 0

        # Update snake body segments
        if len(self.body) > 0:
            for i in range(len(self.body) - 1, 0, -1):
                self.body[i].x = self.body[i - 1].x
                self.body[i].y = self.body[i - 1].y
            self.body[0].x = self.x
            self.body[0].y = self.y

    def draw(self, screen):
        pygame.draw.rect(screen, GREEN, (self.x, self.y, TILE_SIZE, TILE_SIZE))
        for segment in self.body:
            pygame.draw.rect(screen, GREEN, (segment.x, segment.y, TILE_SIZE, TILE_SIZE))

# Food class
class Food:
    def __init__(self):
        self.x = TILE_SIZE * 10
        self.y = TILE_SIZE * 10

    def draw(self, screen):
        pygame.draw.rect(screen, RED, (self.x, self.y, TILE_SIZE, TILE_SIZE))

def place_food():
    food.x = random.randint(0, COLS - 1) * TILE_SIZE
    food.y = random.randint(0, ROWS - 1) * TILE_SIZE
    while (food.x, food.y) in [(snake.x, snake.y)] + [(segment.x, segment.y) for segment in snake.body]:
        food.x = random.randint(0, COLS - 1) * TILE_SIZE
        food.y = random.randint(0, ROWS - 1) * TILE_SIZE

def heuristic(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

def find_path(start, goal, body):
    # A* algorithm
    open_set = []
    heapq.heappush(open_set, (0, start))
    came_from = {}
    g_score = {start: 0}
    f_score = {start: heuristic(start, goal)}

    while open_set:
        _, current = heapq.heappop(open_set)

        if current == goal:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.reverse()
            return path

        for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            neighbor = (current[0] + dx * TILE_SIZE, current[1] + dy * TILE_SIZE)

            if neighbor[0] < 0 or neighbor[0] >= WINDOW_WIDTH or neighbor[1] < 0 or neighbor[1] >= WINDOW_HEIGHT:
                continue

            if neighbor in [(segment.x, segment.y) for segment in body]:
                continue

            tentative_g_score = g_score[current] + 1

            if neighbor not in g_score or tentative_g_score < g_score[neighbor]:
                came_from[neighbor] = current
                g_score[neighbor] = tentative_g_score
                f_score[neighbor] = g_score[neighbor] + heuristic(neighbor, goal)
                heapq.heappush(open_set, (f_score[neighbor], neighbor))

    return []

def is_safe_path(path, body):
    future_body = [Tile(snake.x, snake.y)] + snake.body[:-1]
    for step in path:
        if step in [(segment.x, segment.y) for segment in future_body]:
            return False
        future_body.insert(0, Tile(step[0], step[1]))
        future_body.pop()
    return True

def find_safe_path(start, goal, body):
    path = find_path(start, goal, body)
    if path and is_safe_path(path, body):
        return path
    return find_path_to_tail(body)

def find_path_to_tail(body):
    tail = (body[-1].x, body[-1].y)
    return find_path((snake.x, snake.y), tail, body)

# Initialize game variables
snake = Snake()
food = Food()
place_food()
snake.path = find_safe_path((snake.x, snake.y), (food.x, food.y), snake.body)

# Set up the display
screen = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption('Snake Game')

clock = pygame.time.Clock()

# Main game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    snake.move()

    screen.fill(BLACK)
    snake.draw(screen)
    food.draw(screen)

    # Display score
    font = pygame.font.SysFont('Arial', 20)
    score_text = font.render(f'Score: {snake.score}', True, WHITE)
    screen.blit(score_text, (10, 10))

    # Display game over message
    if snake.game_over:
        game_over_text = font.render(f'Game Over: {snake.score}', True, WHITE)
        screen.blit(game_over_text, ((WINDOW_WIDTH - game_over_text.get_width()) // 2, (WINDOW_HEIGHT - game_over_text.get_height()) // 2))

    pygame.display.flip()
    clock.tick(60)

# Quit pygame properly
pygame.quit()
```
